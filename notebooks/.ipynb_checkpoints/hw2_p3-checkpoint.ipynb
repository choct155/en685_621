```python
import numpy as np
import pandas as pd
from typing import Sequence

project_dir = "/home/choct155/projects/math/algorithms/algoforDS/en685_621/"
data = dict(
    confirmed = pd.read_csv("{}data/time_series_covid_19_confirmed_US.csv".format(project_dir)),
    deaths = pd.read_csv("{}data/time_series_covid_19_deaths_US.csv".format(project_dir)),
    recovered = pd.read_csv("{}data/datasets_494724_1273275_time_series_covid_19_recovered.csv".format(project_dir)),
)

data["confirmed"].head(20)
```

```python
def convert_to_long(data: pd.DataFrame, drop_cols: Sequence[str], idx_cols: Sequence[str]) -> pd.DataFrame:
    long_data = pd.DataFrame(
    data.drop(drop_cols, axis=1)
        .set_index(idx_cols)
        .stack()
    ).reset_index()
    long_data.columns = [s.lower() for s in idx_cols] + ["date", "count"]
    us_data: pd.DataFrame = long_data[~long_data["admin2"].isnull()]
    us_data.columns = ["city", "state", "country", "date", "count"]
    return us_data

test_long = convert_to_long(
    data["confirmed"], 
    ["UID", "iso2", "iso3", "code3", "FIPS", "Lat", "Long_", "Combined_Key"],
    ["Admin2", "Province_State", "Country_Region"]
)
test_long
```

```python
def series_by_city(data: pd.DataFrame, state: str) -> pd.DataFrame:
    agg_by_city: pd.DataFrame = (
        data
        .groupby(["city", "state", "date"])
        .sum()
        .reset_index()
    )
    agg_by_city.columns = ["city", "state", "date", "count"]
    agg_by_city["date"] = pd.to_datetime(agg_by_city["date"])
    city_series: pd.DataFrame = (
        agg_by_city[agg_by_city["state"] == state]
            .set_index(["date", "city"])

    )
    city_series: pd.DataFrame = city_series["count"].unstack("city")
    return city_series

series_by_city(test_long, "Alabama")
```

```python
def series_by_state(data: pd.DataFrame) -> pd.DataFrame:
    agg_by_state: pd.DataFrame = (
        data
        .groupby(["state", "date"])
        .sum()
        .reset_index()
    )
    agg_by_state.columns = ["state", "date", "count"]
    agg_by_state["date"] = pd.to_datetime(agg_by_state["date"])
    state_series: pd.DataFrame = agg_by_state.set_index(["date", "state"]).unstack("state")
    state_series.columns = [col[1] for col in state_series.columns]
    return state_series

series_by_state(test_long)
```

```python
def to_city_series(data: pd.DataFrame, drop_cols: Sequence[str], idx_cols: Sequence[str], state: str) -> pd.DataFrame:
    long_data: pd.DataFrame = convert_to_long(data, drop_cols, idx_cols)
    return series_by_city(long_data, state)

test_city = to_city_series(
    data["confirmed"], 
    ["UID", "iso2", "iso3", "code3", "FIPS", "Lat", "Long_", "Combined_Key"],
    ["Admin2", "Province_State", "Country_Region"],
    "Alabama"
)
test_city
```

```python
def to_state_series(data: pd.DataFrame, drop_cols: Sequence[str], idx_cols: Sequence[str]) -> pd.DataFrame:
    long_data: pd.DataFrame = convert_to_long(data, drop_cols, idx_cols)
    return series_by_state(long_data)

test_state = to_state_series(
    data["confirmed"], 
    ["UID", "iso2", "iso3", "code3", "FIPS", "Lat", "Long_", "Combined_Key"],
    ["Admin2", "Province_State", "Country_Region"]
)
test_state
```

```python
def process_recovered(data: pd.DataFrame) -> pd.DataFrame:
    us_data: pd.DataFrame = (
        data[data["Country/Region"] == "US"]
            .set_index(["Province/State", "Country/Region", "Lat", "Long"])
            .stack()
            .reset_index()
    )
    us_data.columns = ["state", "country", "lat", "long", "date", "count"]
    us_data["date"] = pd.to_datetime(us_data["date"])
    return us_data[["date", "count"]].set_index("date")

process_recovered(data["recovered"])
```

```python
from collections import namedtuple
from typing import NamedTuple, List
from numbers import Rational
```

```python
x = np.arange(10)
xcs = np.cumsum(x)

class CaseStats(NamedTuple):
    stock: Rational
    flow: Rational
    max_flow: Rational
    avg_flow: Rational
    onset_days: int


def fold_stats(data: List[Rational]) -> List[CaseStats]:

    def loop(data: List[Rational], agg: List[CaseStats]=[]) -> List[CaseStats]:
        if len(data) == 0:
            return agg
        elif len(agg) == 0:
            first: CaseStats = CaseStats(data[0], data[0], data[0], data[0], 0)
            return loop(data[1:], [first])
        else:
            last_stat: CaseStats = agg[-1]
            next_stock: Rational = data[0]
            next_flow: Rational = next_stock - last_stat.stock
            next_max_flow: Rational = next_flow if next_flow > last_stat.max_flow else last_stat.max_flow
            next_onset_days: int = (
                0 if next_max_flow == 0
                else last_stat.onset_days + 1
            )
            next_avg_flow: Rational = (
                0 if next_onset_days == 0
                else (((next_onset_days - 1) * last_stat.avg_flow) + next_flow) / next_onset_days
            )
            next_stat: CaseStats = CaseStats(
                next_stock,
                next_flow,
                next_max_flow,
                next_avg_flow,
                next_onset_days
            )
            return loop(data[1:], agg + [next_stat])
    return loop(data)


pd.DataFrame(fold_stats(xcs))
# fold_stats(xcs)
```

```python
class CaseStatsByGeo(NamedTuple):
    geo: str
    stock: Rational
    flow: Rational
    max_flow: Rational
    avg_flow: Rational
    onset_days: int

def max_mean_by_area(data: pd.DataFrame) -> pd.DataFrame:
    geos: Sequence[str] = data.columns
    def get_stats(data: np.array, geo: str) -> CaseStatsByGeo:
        stats: List[CaseStats] = fold_stats(data)
        last_stat: CaseStats = stats[-1]
        return CaseStatsByGeo(
            geo, 
            last_stat.stock, 
            last_stat.flow, 
            last_stat.max_flow, 
            last_stat.avg_flow, 
            last_stat.onset_days
        )
    stats_by_geo: List[CaseStatsByGeo] = [get_stats(data[geo], geo) for geo in geos]
    return pd.DataFrame(stats_by_geo)
    
max_mean_by_area(test_state)
```
